# Swipe plot with hvplot

This example demonstrates how to create a swipe plot with hvplot and panel. The swipe plot allows you to compare two plots by swiping a slider to reveal one plot or the other. In this example, we create two line plots using hvplot and then use the panel library to create a swipe plot that allows you to compare the two plots.

The hvplot library is a high-level plotting API for the PyData ecosystem built on HoloViews. It provides a high-level API for creating complex, interactive plots with a simple, concise syntax. 

```python
import pandas as pd
import hvplot.pandas
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({'a': np.random.randn(100), 'b': np.random.randn(100)}) 
```

```python
fig_1 = df['a'].hvplot(c='red')
```

```python
fig_2 = df['b'].hvplot(c='blue')
```

```python
import panel as pn
pn.extension()

pn.Swipe(fig_1, fig_2, value=51).servable()
```

# Testing the Flask REST API

This notebook demonstrates how to interact with a **Flask REST API** using Python's `requests` library. We will exercise the main **CRUD** operations (Create, Read, Update, Delete) on an IoT "things" resource.

## Prerequisites

Before running this notebook, make sure that:

1. The Flask API server is **running** (e.g., execute `python app.py` in a terminal).
2. The server is listening on `http://localhost:8080`.
3. The `requests` library is installed (`pip install requests`).

## Setup

We start by importing the `requests` library and defining the **base URI** for the API endpoint.

```python
import requests

# Base URI for the IoT things API
uri = 'http://localhost:8080/iot/api/v1.0/things'
print(f"Base URI: {uri}")
```

---

## 1. GET — Retrieve All Things

The **GET** method is used to retrieve data from the server. Here, we request **all** IoT things available in the API.

- **HTTP Method:** `GET`
- **Endpoint:** `/iot/api/v1.0/things`
- **Expected Response:** A JSON list of all things.

```python
print(f"GET {uri}")
response = requests.get(uri)

print(f"Status Code: {response.status_code}")
print(f"Response:\n{response.text}")
```

---

## 2. GET — Retrieve a Single Thing

We can also retrieve a **specific** thing by appending its **ID** to the URI.

- **HTTP Method:** `GET`
- **Endpoint:** `/iot/api/v1.0/things/1`
- **Expected Response:** A JSON object with the details of thing with `id=1`.

```python
thing_id = 1
print(f"GET {uri}/{thing_id}")
response = requests.get(f"{uri}/{thing_id}")

print(f"Status Code: {response.status_code}")
print(f"Response:\n{response.text}")
```

---

## 3. POST — Create a New Thing

The **POST** method is used to **create** a new resource on the server. We send a JSON payload describing the new IoT thing, including its name, location, and associated sensors.

- **HTTP Method:** `POST`
- **Endpoint:** `/iot/api/v1.0/things`
- **Payload:** A JSON object with `name`, `local`, and `sensors` fields.
- **Expected Response:** The newly created thing (usually with an assigned `id`).

```python
# Define the data for the new IoT thing
data = {
    "local": "@lab. 163 / ISE /UAlg",
    "name": "Apollo",
    "sensors": [
        {
            "sensor_name": "temperature",
            "units": "numerical"
        },
        {
            "sensor_name": "humidity",
            "units": "percent"
        }
    ]
}

print("Data to send:", data)
```

```python
# Send the POST request with the JSON payload
print(f"POST {uri}")
response = requests.post(uri, json=data)

print(f"Status Code: {response.status_code}")
print(f"Response:\n{response.text}")
```

---

## 4. PUT — Update an Existing Thing

The **PUT** method is used to **update** an existing resource. Here we update the name of the thing with `id=3`.

- **HTTP Method:** `PUT`
- **Endpoint:** `/iot/api/v1.0/things/3`
- **Payload:** A JSON object with the fields to update.
- **Expected Response:** The updated thing.

```python
thing_id = 3
update_data = {"name": "Apollo X1"}

print(f"PUT {uri}/{thing_id}")
response = requests.put(f"{uri}/{thing_id}", json=update_data)

print(f"Status Code: {response.status_code}")
print(f"Response:\n{response.text}")
```

### Verify the Update

Let's retrieve all things again to confirm that the update was applied.

```python
print(f"GET {uri}")
response = requests.get(uri)

print(f"Status Code: {response.status_code}")
print(f"All things after PUT:\n{response.text}")
```

---

## 5. DELETE — Remove a Thing

The **DELETE** method is used to **remove** a resource from the server.

- **HTTP Method:** `DELETE`
- **Endpoint:** `/iot/api/v1.0/things/3`
- **Expected Response:** A confirmation that the resource was deleted.

```python
thing_id = 3

print(f"DELETE {uri}/{thing_id}")
response = requests.delete(f"{uri}/{thing_id}")

print(f"Status Code: {response.status_code}")
print(f"Response:\n{response.text}")
```

### Verify the Deletion

Let's retrieve all things one more time to confirm the deletion.

```python
print(f"GET {uri}")
response = requests.get(uri)

print(f"Status Code: {response.status_code}")
print(f"All things after DELETE:\n{response.text}")
```

---

## Summary

In this notebook we tested the four main HTTP methods against our Flask REST API:

| HTTP Method | Operation | Description |
|-------------|-----------|-------------|
| `GET`       | **Read**  | Retrieve all things or a single thing by ID |
| `POST`      | **Create**| Add a new thing with a JSON payload |
| `PUT`       | **Update**| Modify an existing thing by ID |
| `DELETE`    | **Delete**| Remove a thing by ID |

These operations form the basis of any **RESTful API** and map directly to the standard **CRUD** (Create, Read, Update, Delete) operations.
